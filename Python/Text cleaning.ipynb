```python
def remove(text):
    # remove numbers, special characters, tab
    text = re.sub('[^\w&\s.-]|[0-9\t]','',text)
    text = re.sub('\s+',' ',text)
    return text
```

```python
processed_document = remove(document).lower()
```

```python
from nltk.corpus import stopwords
tokenized_sentence = [[word for word in sentence.strip().split(" ") if word not in stopwords.words("english")] for sentence in processed_document.split(".") if sentence]
```
