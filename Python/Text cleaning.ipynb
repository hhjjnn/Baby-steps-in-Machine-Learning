```python
multiWordTerms = ['cost of living adjustments', 'cost of living adjustment']
def replaceMultiWordTerms(text, multiWordTerms):
    # This function takes a string and a list of multiword terms and replaces each occurrence
    # a multiword term with the term having spaces replaced by '_' e.g. Cost Of Living Adjustment
    # will be replaced with Cost_Of_Living_Adjustment
    for term in multiWordTerms:

        text = text.replace(term, term.replace(' ', '_'))

    return text

def remove(text):
    # remove numbers, special characters, tab
    text = re.sub('[^\w&\s.-]|[0-9\t]','',text)
    text = re.sub('\s+',' ',text)
    return text
    

def process(text):
    text = remove(text).lower()
    text = replaceMultiWordTerms(text, multiWordTerms)
    tokenized_sentence = [[word for word in sentence.strip().split(" ") if word not in stopwords.words("english")] for sentence in text.split(".") if sentence]
    return tokenized_sentence
```
